```python
import pandas as pd
```

```python
data=pd.read_csv("Placement.csv")
data
```

```python
data.isnull().sum()
```

```python
data=pd.get_dummies(data,dtype=int,drop_first=True)
```

```python
data
```

```python
data['salary'].fillna(data['salary'].mode()[0],inplace=True)
```

```python
data.isna().sum()
```

```python
data.shape
```

```python
data.columns
```

```python
indep=data[[ 'ssc_p', 'hsc_p', 'degree_p', 'etest_p', 'mba_p', 'status_Placed',
            'gender_M', 'ssc_b_Others', 'hsc_b_Others', 'hsc_s_Commerce',
            'hsc_s_Science', 'degree_t_Others', 'degree_t_Sci&Tech', 'workex_Yes',
            'specialisation_Mkt&HR']]
indep
```

```python
dep=data[['salary']]
dep
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(indep,dep,test_size=0.30,random_state=0)
```

```python
from sklearn.feature_selection import SelectKBest,f_regression
from sklearn.feature_selection import chi2
#from sklearn.frature_selection import RFE
sk=SelectKBest(score_func=f_regression, k=5)
fite=sk.fit_transform(X_train,Y_train)
fite1=sk.transform(X_test)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
```

```python
import pickle
file="stand1.sav"
pickle.dump([sc],open(file,"wb"))
res1=pickle.load(open("stand1.sav","rb"))
```

```python
from sklearn.svm import SVR
regressor=SVR()
regressor.fit(fite,Y_train)
```

```python
y_pred=regressor.predict(fite1)
```

```python
from sklearn.metrics import r2_score
r2score=r2_score(Y_test,y_pred)
r2score
```
